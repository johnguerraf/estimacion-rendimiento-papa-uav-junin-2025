```python
import pandas as pd
import numpy as np
from pathlib import Path
import re

RESULTADO_DIR = Path(r"E:\proyecto_agric_precision\Dataset de Imagenes\ortomosaicos_canales\resultado")
OUT9 = RESULTADO_DIR / "PASO_09_temporal_3fechas_obsUnitId"
DATA_WIDE = OUT9 / "dataset_wide_3dates.csv"

MEAS_DIR = Path(r"E:\proyecto_agric_precision\Dataset de Imagenes\ortomosaicos_canales\measurements")
YIELD_FILE = MEAS_DIR / "Solynta_field_2025_GT_yield.csv"
ADD_FILE = MEAS_DIR / "Solynta_field_2025_GT_additional.csv"  # opcional

df_wide = pd.read_csv(DATA_WIDE)

print("✅ features wide:", df_wide.shape)
print("✅ yield exists:", YIELD_FILE.exists(), YIELD_FILE)
print("✅ additional exists:", ADD_FILE.exists(), ADD_FILE)
df_wide.head()
```

```python
y = pd.read_csv(YIELD_FILE)
print("✅ yield cargado:", y.shape)
print("Columnas yield:", list(y.columns))
display(y.head())
```

```python
def pick_col(cols, patterns):
    for pat in patterns:
        for c in cols:
            if re.search(pat, str(c), re.IGNORECASE):
                return c
    return None

id_col = pick_col(y.columns, [r"plot_id", r"plot", r"obsunitid", r"obsUnitId", r"^id$", r"unit"])
yield_col = pick_col(y.columns, [r"yield", r"rend", r"cosech", r"produ", r"kg", r"t/ha"])

print("Detectado id_col:", id_col)
print("Detectado yield_col:", yield_col)

if id_col is None or yield_col is None:
    raise ValueError("No pude detectar ID o YIELD. Pega aquí las columnas del CSV y lo ajusto.")

y2 = y[[id_col, yield_col]].copy()
y2 = y2.rename(columns={id_col: "id_raw", yield_col: "yield"})
y2["yield"] = pd.to_numeric(y2["yield"], errors="coerce")

# extraer plot_id numérico si el id viene como texto tipo 'plot_12'
def extract_plot_id(val):
    if pd.isna(val): 
        return np.nan
    m = re.search(r"(\d+)", str(val))
    return int(m.group(1)) if m else np.nan

y2["plot_id"] = y2["id_raw"].apply(extract_plot_id)
y2["obsUnitId"] = y2["id_raw"].astype(str)

y2 = y2.dropna(subset=["yield"])
print("✅ y2 limpio:", y2.shape)
display(y2.head())
```

```python
use_plot = y2["plot_id"].notna().mean() > 0.5
print("¿Yield trae plot_id usable?:", use_plot)

if (not use_plot) and ADD_FILE.exists():
    add = pd.read_csv(ADD_FILE)
    print("✅ additional cargado:", add.shape)
    print("Columnas additional:", list(add.columns))
    display(add.head())

    # intenta detectar columnas en additional
    add_id = pick_col(add.columns, [r"obsunitid", r"obsUnitId", r"^id$", r"unit"])
    add_plot = pick_col(add.columns, [r"plot_id", r"plot"])

    print("Detectado add_id:", add_id)
    print("Detectado add_plot:", add_plot)

    if add_id and add_plot:
        add2 = add[[add_id, add_plot]].copy().rename(columns={add_id: "obsUnitId", add_plot: "plot_id"})
        add2["obsUnitId"] = add2["obsUnitId"].astype(str)
        add2["plot_id"] = pd.to_numeric(add2["plot_id"], errors="coerce")

        y2 = y2.merge(add2, on="obsUnitId", how="left", suffixes=("","_mapped"))
        # si plot_id venía vacío, usa el mapeado
        y2["plot_id"] = y2["plot_id"].fillna(y2.get("plot_id_mapped"))
        use_plot = y2["plot_id"].notna().mean() > 0.5

        print("✅ Tras mapear, ¿plot_id usable?:", use_plot)
    else:
        print("⚠️ additional no tiene columnas claras para mapear.")
```

```python
OUT10 = RESULTADO_DIR / "PASO_10_yield_2025"
OUT10.mkdir(parents=True, exist_ok=True)

if use_plot:
    yield_by = (y2.dropna(subset=["plot_id"])
                .groupby("plot_id")["yield"]
                .mean()
                .reset_index()
                .sort_values("plot_id"))
    join_key = "plot_id"
else:
    yield_by = (y2.groupby("obsUnitId")["yield"]
                .mean()
                .reset_index()
                .sort_values("obsUnitId"))
    join_key = "obsUnitId"

out_yield = OUT10 / "yield_by_plot_2025.csv"
yield_by.to_csv(out_yield, index=False)

print("✅ Guardado:", out_yield)
print("✅ Join key para Paso 11:", join_key)
display(yield_by.head())
```
