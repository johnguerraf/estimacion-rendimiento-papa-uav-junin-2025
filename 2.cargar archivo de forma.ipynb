```python
import geopandas as gpd

# 1) Ruta del shapefile (usa raw string en Windows)
shapefile_path = r"E:\proyecto_agric_precision\Dataset de Imagenes\ortomosaicos_canales\metadata\plot_shapefile.shp"

# 2) Leer shapefile
gdf = gpd.read_file(shapefile_path)

# 3) Ver CRS
print("Shapefile CRS:", gdf.crs)

# 4) Asegurar CRS correcto (debe ser EPSG:32631)
# Si ya está en 32631, no hace nada. Si no, lo reproyecta.
target_crs = "EPSG:32631"
if gdf.crs is None:
    raise ValueError("El shapefile no tiene CRS definido. Asígnalo en QGIS y vuelve a intentar.")
elif gdf.crs.to_string() != target_crs:
    gdf = gdf.to_crs(target_crs)
    print("Reproyectado a", target_crs)

# 5) Crear plot_id (tu shapefile NO tiene 'plot_id', pero sí 'obsUnitId')
if "obsUnitId" in gdf.columns:
    gdf["plot_id"] = gdf["obsUnitId"]
elif "new_id" in gdf.columns:
    gdf["plot_id"] = gdf["new_id"]
elif "new_id2" in gdf.columns:
    gdf["plot_id"] = gdf["new_id2"]
else:
    # Último recurso: crear IDs automáticos
    gdf["plot_id"] = range(1, len(gdf) + 1)

# 6) Verificar que plot_id exista y sea único
print("Columnas disponibles:", list(gdf.columns))
print("N parcelas:", len(gdf))
print("plot_id únicos:", gdf["plot_id"].nunique())

# 7) Mostrar ejemplo
display(gdf[["plot_id", "geometry"]].head())

# 8) (Opcional) Exportar atributos a Excel
excel_file_path = r"E:\proyecto_agric_precision\Dataset de Imagenes\ortomosaicos_canales\resultado\shape2025att_12_10_2025.xlsx"
gdf.to_excel(excel_file_path, index=False)
print("Exportado a:", excel_file_path)
```
